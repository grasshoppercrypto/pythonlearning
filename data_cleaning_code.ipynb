```python
import pandas as pd

# Read the online file by the URL provides above, and assign it to variable "df"
other_path = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df = pd.read_csv(other_path, header=None)
```

```python
df.head()
#print the first 5 row and found the data might needs some cleaning... for example: ?, column names, etc
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# add header column names
df.columns = headers
df.head(10)
```

```python
# price and normalized-losses columns has questions marks, need to drop the null data
df=df.dropna(subset=["price"], axis=0)
#df=df.dropna(subset=["normalized-losses"], axis=0) #this line didn't seem to work,. need to find out why?

# it could convert all the ? into Nah value and try drom them again, use numpy method np.nan
df.replace("?", np.nan, inplace = True)
df.tail(20)

# now there should be some discussion about how to deal with missing data
# first let's do a stat of missing data
missing_data = df.isnull()
missing_data.head(10)
```

```python
df.describe()
```

```python
# to understand the full stats of missing data groub by columns, here is the code to do that
# use a for loop to sweep all columns
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
df.dtypes #it shows that several columns types are object, but it should be either float or int, like normalized-losses, price, etc
```

```python
# lets do several different data cleaning:
# replace the mising data in those columns (normalized-losses, bore, stroke, horsepower, peak-rpm) by mean
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
# avg_norm_loss = df["normalized-losses"].mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

avg_bore = df["bore"].astype("float").mean(axis=0)
# avg_norm_loss = df["normalized-losses"].mean(axis=0)
print("Average of Bore:", avg_norm_loss)
df["bore"].replace(np.nan, avg_norm_loss, inplace=True)

avg_stroke=df['stroke'].astype('float').mean(axis=0)
print('Average of stroke',avg_stroke)
df['stroke'].replace(np.nan, avg_stroke, inplace=True)

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)

#now check the null stats again, there is still number of doors need to be cleaned!
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
# to understand the stats of num of doors column
df['num-of-doors'].value_counts()
df['num-of-doors'].value_counts().idxmax()
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)

df.reset_index(drop=True, inplace=True)
```

```python
#this only shows the stat results of numerical columns
# still something is not right, those filled columns are not showing
# probably need to change the type of those columns from object to float or int...
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df["horsepower"]=df["horsepower"].astype(int, copy=True)
df.describe() 
# looks much better now
```
